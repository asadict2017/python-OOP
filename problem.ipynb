Problem 1
Write a program that prompts the user to enter a number and determines whether it's positive, negative, or zero using if-else conditions.

```python
num=int(input("please input number"))
#condition
if num>0:
  print(f"Input number {num} is positive")
elif num<0:
  print(f"Input number {num} is negative")
else:
  print(f"Input number {num} is zero")
```

Problem 2
Create a program that asks for a person's age and then categorizes them as a child, teenager, adult, or senior based on the age range as follows

child: age < 13

teenager: 13 to 19 inclusive

adult: 18 or above but under 65

senior: 65 or older




```python
age=int(input("enter your age"))
if age<13:
  print("child")
elif age>=13 and age<=19:
  print("teenager")
elif age>=18 and age<65:
  print("adult")
else:
  print("senior")

```

```python
#alternative
age=int(input("enter age in number"))
if age<=12:
  print("child")
elif age<=19:
  print("teenager")
elif age<=64:
  print("adult")
else:
  print("senior")
```

Write a Python program that checks whether a given year is a leap year or not. Use an if-else statement to determine this.

The logic for checking if a year is a leap year is as follows:

If the year is divisible by 4 (year % 4 == 0) and not divisible by 100 (year % 100 != 0), it's a leap year.

If the year is divisible by 400, it's also a leap year.

Using this logic, you can determine whether a given year is a leap year or not.

```python
year_input=int(input("enter year: "))
if (year_input%4==0 and year_input%100 !=0) or year_input%400==0:
  print("leap year")
else:
  print("not leap year")
```
