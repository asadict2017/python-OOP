```python
#problem : create a car class and define a constructor of that class and also define a method that will return some attribute info and finally create object of that class and call attribute/method of that class

class car:
  #defining constructor
  def __init__(self,cartype,brand,manufacturingYr):
    self.cartype=cartype
    self.brand=brand
    self.manufacturingYr=manufacturingYr
  #defining method
  def carinfo(self):
    return(f"{self.cartype},brand{self.brand},year of manufacturing {self.manufacturingYr} ")
#creating object by creating constructor of that class
car1=car("Truck","Toyota","2020")
new_car=car1.carinfo()
print(new_car)
```
